# Model

```python
import torch
import echotorch
from echotorch.datasets.NARMADataset import NARMADataset
import echotorch.nn.reservoir as etrs
# import echotorch.utils.matrix_generation as mg
from torch.autograd import Variable
from torch.utils.data.dataloader import DataLoader
import numpy as np
import matplotlib.pyplot as plt
import json

from utils import *

def build_model(prediction_horizon, params):

    # Manual seed initialisation
    np.random.seed(1)
    torch.manual_seed(1)

    # Reservoir hyper-parameters
    input_dim = 1
    output_dim = 1

    spectral_radius = params['spectral_radius']
    reservoir_size = int(params['reservoir_size'])
    connectivity = params['connectivity']
    ridge_param = params['ridge_param']
    input_scaling = params['input_scaling']
    bias_scaling = params['bias_scaling']

    batch_size = 1 

    # Internal matrix
    w_generator = echotorch.utils.NormalMatrixGenerator(
        connectivity=connectivity,
        spetral_radius=spectral_radius
    )

    # Input weights
    win_generator = echotorch.utils.NormalMatrixGenerator(
        connectivity=connectivity,
        scale=input_scaling,
        apply_spectral_radius=False
    )

    # Bias vector
    wbias_generator = echotorch.utils.NormalMatrixGenerator(
        connectivity=connectivity,
        scale=bias_scaling,
        apply_spectral_radius=False
    )

    esn = etrs.ESN(
        input_dim=input_dim,
        hidden_dim=reservoir_size,
        output_dim=output_dim,
        learning_algo='inv',
        w_generator=w_generator,
        win_generator=win_generator,
        wbias_generator=wbias_generator,
        ridge_param=ridge_param
    )
    # data parameters
#     prediction_horizon = 5

    ds_builder = DatasetBuilder()
    X_train, y_train, X_test, y_test = ds_builder.build_dataset_simple(prediction_horizon, interval, data_path='../data/input_data.csv', train_pct=0.8)

    dataset_train = MyDataset(X_train, y_train)
    dataset_test = MyDataset(X_test, y_test)

    trainloader = DataLoader(dataset_train, batch_size=batch_size, shuffle=False)
    testloader = DataLoader(dataset_test, batch_size=batch_size, shuffle=False)

    for data in trainloader:
        # Inputs and outputs
        inputs, targets = data

        # Transform data to Variables
        inputs, targets = Variable(inputs), Variable(targets)

        # ESN need inputs and targets
        esn(inputs, targets)
    # Now we finalize the training by

    # computing the output matrix Wout.
    esn.finalize()

    # Test MSE
    dataiter = iter(testloader)
    test_u, test_y = dataiter.next()
    test_u, test_y = Variable(test_u), Variable(test_y)
    y_predicted = esn(test_u)
    y_predicted = ds_builder.unscale_y(y_predicted.data)
    test_y = ds_builder.unscale_y(test_y.data)
#     print(u"Test RMSE: {}".format(echotorch.utils.rmse(y_predicted, test_y)))
    
    return float(echotorch.utils.rmse(y_predicted, test_y))
```

# Parameters vs RMSE plots

```python
from collections import defaultdict
import matplotlib.pyplot as plt


def plot(results):
    for param in list(results.keys()):
        fig = plt.figure(figsize=(10, 6))
        plt.plot(results[param]['param'], results[param]['rmse'])
        if param == 'ridge_param':
            plt.xscale('log')
        plt.xlabel('Value')
        plt.ylabel('RMSE')
        plt.title(param)
        plt.show()
        
        
def get_best_params(interval='1min'):
    # Read best hyperparams from JSON
    with open(f'hyperparams/hyperparams_esn_{interval}.json', 'r') as fp:
        default_params = json.load(fp)

    params = dict()
    params['spectral_radius'] = default_params['spectral_radius']
    params['reservoir_size'] = int(default_params['reservoir_size'])
    params['connectivity'] = default_params['connectivity']
    params['ridge_param'] = default_params['ridge_param']
    params['input_scaling'] = default_params['input_scaling']
    params['bias_scaling'] = default_params['bias_scaling']
    
    return params

param_ranges = dict()
param_ranges['spectral_radius'] = np.arange(0.1, 20, 4)
param_ranges['reservoir_size'] = np.arange(50, 500, 50)
param_ranges['connectivity'] = np.arange(0.1, 1.0, 0.1)
param_ranges['ridge_param'] = np.logspace(-10, 2, base=10, num=10)
param_ranges['input_scaling'] = np.arange(0.1, 1.1, 0.1)
param_ranges['bias_scaling'] = np.arange(0.0, 1.1, 0.1)

param_names = list(get_best_params().keys())

prediction_horizons = [5, 8, 7]
interval = ['1min', '60min', 'daily']

for prediction_horizon, interval in zip(prediction_horizons, interval):
    print(f'Interval: {interval}, prediction horizon: {prediction_horizon}')
    results = {param:{'rmse':[], 'param':list(param_ranges[param])} for param in param_names}

    for param in param_names:
        rmse = []
        for param_value in param_ranges[param]:
            best_params = get_best_params(interval)
            best_params[param] = param_value

            rmse_ = build_model(prediction_horizon, best_params)
            rmse.append(rmse_)

        results[param]['rmse'] = rmse
        
    plot(results)
```
