```python
import pandas as pd
from pandas import read_csv
from pandas import datetime
from pandas import DataFrame
from statsmodels.tsa.arima_model import ARIMA
from matplotlib import pyplot
from sklearn.metrics import mean_squared_error
import csv
```

```python
col_list = ["date", "close"]
series = pd.read_csv('../data/input_data.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, usecols=col_list).iloc[::-1].reset_index(drop=True)
X = series.values
size = int(len(X) * 0.8)
train, test = X[0:size], X[size:len(X)]
realValues = [x for x in train]
predictions = list()
result = list()
mycsv = csv.writer(open('results.csv', 'wb'))

for t in range(len(test)):
    model = ARIMA(realValues, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    prediction = output[0]
    predictions.append(prediction)
    expected = test[t]
    realValues.append(expected)
    result.append
    # print('predicted=%f, expected=%f' % (prediction, expected))

    
error = mean_squared_error(test, predictions, squared=False)
print('Test RMSE: %.3f' % error)
# plot

df = pd.DataFrame(predictions)
df.to_csv('predictions.csv')

pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```
